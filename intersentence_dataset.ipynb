```python
import json
j=open("/home/judson/Desktop/inter_06_7.json","r").readlines()
dct=[]
for i in j:
    dct.append(json.loads(i))
```

```python
def dataset_creatot(dct,result_dict):
    for i,j in enumerate(dct):
        if len(dct)!=i+1:
            if dct[i]["doc_id"]==dct[i+1]["doc_id"]-1:
                key=dct[i]["labels"][0]+"/"+dct[i+1]["labels"][0]
                if key not in result_dict:
                    result_dict[key]=[[dct[i]['text'],dct[i+1]['text']]]
                else:
                    result_dict[key].append([dct[i]['text'],dct[i+1]['text']])
    return result_dict
```

```python
result_dict=dataset_creatot(dct,result_dict)
```
